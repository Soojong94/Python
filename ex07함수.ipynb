### 함수
- 호출될 때 특정 기능을 실행하는 재사용 가능한 코드 블록
- 지정된 코드에 이름을 붙여 관리하는 방법
- 코드의 간결함을 높여주고, 재사용성을 높여준다

- 1. 내장 함수 2. 사용자정의 함수
```python
def 함수명 (매개변수) :
    실행문장
    return 반환변수
```
- 함수호출 :  함수명(인수)



````python
# 함수 정의
def number_sum(n1, n2) :
    result = n1 + n2
    return result
````

#### 3. 파이썬함수 특징

````python
# 함수에서 선언된 변수는 지역변수
a = 10 # 전역변수 선언
def test() :
    a = 1 # 지역변수로 인식
    print(a)
test()
print(a)
````

````python
# global 키워드를 이용해 전역변수로 인식시키자

a = 10 # 전역변수 선언

def test() :
    global a
    a = 1 # 전역변수로 인식
    
test()
print(a)
````

#### 4. docstring
- 함수에 대한 부가 설명을 작성하는 기능

````python
def cal(num1, num2, op) :
    """ 
    덧셈과 뺄셈을 계산하는 함수 
    op : 연산자를 의미, 주로 '+' 기호를 입력
    """
    if op == '+' :
        return num1 + num2 
    else :
        return num1 - num2
````

````python
cal(10,5,'+') 
````

#### 5. 가변인자
- 사용자가 입력인자를 정해진 길이 없이 넣고 싶은 경우

````python
   # 튜플 스타일

def fun1(name, *score) :
    print("이름 : " , name)
    print("성적 : " , score)
````

````python
fun1("김수종", 100,100,80,2)
````

````python
# 딕셔너리 스타일
def fun2 (name, **score) :
    print("이름 : " , name)
    print("성적 : " , score)
````

````python
fun2("김수종" , java=100, python=80, html=40, css=50)
````

#### 4. 익명함수
- lambda 표현식 : 메소드를 하나의 식으로 표현한 것 일반적으로 짧고 한 줄로 표현할 수 있는 함수에 대해 사용


````python
# 일반적인 함수 정의
def mul(x,y) :
    return x*y
result = mul(3,5)
print(result)
````

````python
# 람다식을 이용한 함수 정의
mul2 = lambda x,y : x*y 
mul2(9,7)
````
